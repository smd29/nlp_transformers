```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
### sentences
sentences=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]

maxLen = 0
for s in sentences:
  maxLen = max(len(s),maxLen)
```

```python
vocabulary_size=1000
```

One hot encoding

```python
one_hot_encoding = [one_hot(words,vocabulary_size) for words in sentences]
one_hot_encoding
```

Word embedding

```python
embedded_docs = pad_sequences(one_hot_encoding,maxlen=maxLen,padding='pre')
embedded_docs
```

```python
features = 10
```

```python
model = Sequential()
model.add(Embedding(vocabulary_size,features,input_length=maxLen))
model.compile('adam','mse')
model.summary()
```

```python
model.predict(embedded_docs)
# each row contains 10(feature size) values
```
